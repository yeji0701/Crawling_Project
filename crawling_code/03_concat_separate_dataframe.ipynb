```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
first = pd.read_csv('./datas/first.csv', index_col=[0])
first.tail()
```

```python
second = pd.read_csv('./datas/second.csv', index_col=[0])
second.tail()
```

```python
third = pd.read_csv('./datas/third.csv', index_col=[0])
third.tail()
```

```python
fourth = pd.read_csv('./datas/fourth.csv', index_col=[0])
fourth.tail()
```

```python
frames = [first, second, third, fourth]
df = pd.concat(frames)
df.drop(columns=['Link'], inplace=True)
```

```python
df.to_csv('./datas/df~2000.csv')
```

```python
fifth = pd.read_csv('./datas/fifth.csv', index_col=[0])
fifth.tail()
```

```python
frame = [df, fifth]
df = pd.concat(frame)
df
```
